```python
%matplotlib notebook
from pyfeti.src.utils import DomainCreator, DiskCreator, DiskCreator_3D, PrismaCreator
import amfe
import numpy as np
import matplotlib.pyplot as plt
```

```python
width = 6
heigh = 4
dobj = DomainCreator(width=width,heigh=heigh,x_divisions=21,y_divisions=11)
mesh_file = 'beam.msh'
dobj.save_gmsh_file(mesh_file)

m1 = amfe.Mesh()
m1.import_msh(mesh_file)

m2 = m1.translation(np.array([width,0.]))

ax1 = amfe.plot2Dmesh(m1)
amfe.plot2Dmesh(m2,ax=ax1)

ax1.set_xlim([0,2*width])
ax1.set_ylim([0,heigh])
ax1.set_aspect('equal')
ax1.set_xlabel('Width [m]')
ax1.set_ylabel('Heigh [m]')
plt.legend('off')
```

```python
r_in_1 = 6
r_out_1 = 18
angle=60
theta_0=(90-angle/2)
angle_divisions = 25
radial_divisions = 16
dobj1 = DiskCreator(r_in=r_in_1,r_out=r_out_1, sector_angle=angle, theta_0=theta_0,angle_divisions=angle_divisions,radial_divisions=radial_divisions)


mesh_file1 = 'mesh_disk.msh'
dobj1.save_gmsh_file(mesh_file1)

m3 = amfe.Mesh()
m3.import_msh(mesh_file1)



fig, ax1 = plt.subplots(1,1,figsize=(7,7))
amfe.plot2Dmesh(m3,ax=ax1)


mult=1.2
for ax in (ax1,):
    ax.set_xlim([-mult*r_out_1,mult*r_out_1])
    ax.set_ylim([-mult*r_out_1,mult*r_out_1])
    ax.set_aspect('equal')
    ax.set_xlabel('Width [m]')
    ax.set_ylabel('Heigh [m]')
plt.legend('off')
```

```python
width,heigh,thickness = 10,2,1
creator_obj  = PrismaCreator(width,heigh,thickness,
                             x_divisions=11,y_divisions=7,z_divisions=4)

mesh_path = 'prisma.msh'
creator_obj.save_gmsh_file(mesh_path)
m4 = amfe.Mesh()
m4.import_msh(mesh_path)

ax5 = amfe.plot3Dmesh(m4)

ax =ax5
mult = 1.2
ax.set_xlim([-mult,mult*width])
ax.set_ylim([-mult,mult*heigh])
ax.set_zlim([-mult,mult*thickness])
ax.set_aspect('equal')
ax.set_xlabel('Width [m]')
ax.set_ylabel('Heigh [m]')
ax.set_zlabel('Thickness [m]')
```

```python
r_in,r_out,thickness=10,30,5
creator_obj  = DiskCreator_3D(r_in,r_out,thickness=thickness,
                                   sector_angle=65,
                                   angle_divisions=11,
                                   radial_divisions=4,
                                   theta_0=0,
                                   start_x=0,
                                   start_y=0)
        
mesh_path = 'disk3d.msh'    
creator_obj.save_gmsh_file(mesh_path)
m5 = amfe.Mesh()
m5.import_msh(mesh_path)

ax6 = amfe.plot3Dmesh(m5)
mult=1.2
ax =ax6
ax.set_xlim([-mult*r_out,mult*r_out])
ax.set_ylim([-mult*r_out,mult*r_out])
ax.set_zlim([-mult*r_out,mult*r_out])
ax.set_aspect('equal')
ax.set_xlabel('Width [m]')
ax.set_ylabel('Heigh [m]')
ax.set_zlabel('Thickness [m]')
plt.legend('off')
```
